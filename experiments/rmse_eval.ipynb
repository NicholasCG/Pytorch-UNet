```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
glare_img = cv2.imread('../data/glares/img2_glare3.png')
```

```python
original_img = cv2.imread('../data/imgs/img2.png')
```

```python
residuals = glare_img - original_img
```

```python
def display_img(image):
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    ax.imshow(image)
    ax.xaxis.tick_top()
```

```python
display_img(glare_img)
display_img(original_img)
display_img(residuals)
```

```python
def mse(x, y):
    return np.mean((x - y) ** 2)
```

```python
# MSE between glare minus residuals and original image
print(mse(glare_img - residuals, original_img))
```

```python
# MSE between glare and original image
print(mse(glare_img, original_img))
```
